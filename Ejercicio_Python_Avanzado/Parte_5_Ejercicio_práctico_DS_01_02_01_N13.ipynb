## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
# veo que hay un 'Un' con mayúscula y otro 'un' con minúscula, Python los va
# a considerar elementos diferentes, así que haré algo para que los considere iguales

ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", \
          "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

def encontrar_indices(lista, palabra):
    indices = []   # lista vacía donde guardaremos los índices
    for i in range(len(lista)):
        if lista[i].lower() == palabra.lower():   # voy a obviar las mayúsculas y minúsculas
            indices.append(i)
    return indices

print(encontrar_indices(ejer_3, "un"))
print(encontrar_indices(ejer_3, "es"))
print(encontrar_indices(ejer_3, "binario"))
`````

`````python
# EXTRA
# Me ha surgido una duda cuando hacía el ejercicio anterior:
# Si fuera la lista mucho más larga y no pudiera ir uno a uno para ver qué palabras son
# las que se repiten, ¿cómo lo haría? para saber cuáles son y luego hacer lo
# que pide el ejercicio

ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", \
          "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

# voy a contar cuántas veces aparece cada palabra (ignorando las mayúsculas)
contador = {}
for palabra in ejer_3:
    p = palabra.lower()
    if p in contador:
        contador[p] += 1
    else:
        contador[p] = 1

print(contador)

# ahora quiero mostrar solo las palabras que están repetidas y sus índices
for palabra, veces in contador.items():
    if veces > 1:   # quiero solo las que aparecen más de una vez
        print(palabra, "aparece en los índices", encontrar_indices(ejer_3, palabra))
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def validar_dni(dni):
    if len(dni) < 8 or len(dni) > 9:
        return False
    
    numero = dni[:-1]
    letra = dni[-1].upper()
    
    if not numero.isdigit():
        return False

    numero = int(numero)

    palabra = 'TRWAGMYFPDXBNJZSQVHLCKE'
    letra_correcta = palabra[numero % 23]
    return letra == letra_correcta

# voy a probar:
print(validar_dni("00000000T"))  # Ejemplo válido
print(validar_dni("12345678Z"))  # Puede ser válido o no según cálculo
print(validar_dni("1234A678Z"))  # Inválido
print(validar_dni("1234567A"))   # Inválido
`````
